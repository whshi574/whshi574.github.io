```python
# Mian Folder search

import requests
import json

# 每次都需要更新
tenant_access_token = "t-g1044g0OFQ2ZMX4UMLMETCYIXHQIMUESU37EUE5C"

# 你所想要文件夹的url
foldertoken = "fldcntNpxLpHDkztVVRjS9kDuQf"



#飞书API请求url，这个链接形式可以在后台进行快速调整
url = "https://open.feishu.cn/open-apis/drive/v1/files?folder_token="+foldertoken
payload = ''

headers = {
  'Authorization': 'Bearer '+tenant_access_token
}

response = requests.request("GET", url, headers=headers, data=payload)

if response.status_code == 200:
    # 解析API响应结果
    result = json.loads(response.text)
    file_list = result['data']['files']
    for file in file_list:
        print(file['name'], file['url'])
else:
    print(f"获取文件列表失败，状态码：{response.status_code}")
    
```
